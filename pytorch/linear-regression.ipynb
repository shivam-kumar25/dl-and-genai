`nn` is the neural network module in PyTorch, providing building blocks for creating and training deep learning models.




```python
import torch 

from torch import nn 


from matplotlib import pyplot as plt 
```

```python
# Generating data - Define ground truth for Linear Regression 

weight = 0.7   # Slope parameter 
bias = 0.3   # Intercept parameter 


# Generate Sysnthetic data with known parameters (weight and bias) 
start = 0 
end = 1
step = 0.02 

X = torch.arange(start, end, step)  # Shape: (50,)
display("Tensor X (When unsqueeze is not used):", X, X.shape)

X = torch.arange(start, end, step).unsqueeze(dim=1)  # Shape: (50, 1)
# here unsqueeze is used to add an extra dimension to the tensor, making it a 2D tensor with shape (50, 1) instead of a 1D tensor with shape (50,). This is often done to ensure that the data is in the correct format for certain operations, such as matrix multiplication or when feeding it into a neural network.
display("Tensor X:", X, X.shape)

y = weight * X + bias  
display("Tensor y:", y, y.shape)


print(f"X shape: {X.shape} | y shape: {y.shape}")       
```

```python
# 

train_split = int(0.8 * len(X))  # 80% for training, 20% for testing

# 
X_train, y_train = X[:train_split], y[:train_split]  

#    
X_test, y_test = X[train_split:], y[train_split:]


# Final check of the data 
print(f"X_train shape: {X_train.shape} | y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape} | y_test shape: {y_test.shape}") 

```
